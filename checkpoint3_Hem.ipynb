```python
import pandas as pd
```

```python
df = pd.read_csv('/content/Youtube Influencer Analysis - Updated.csv',sep=",",encoding='cp1252')

df
```

```python
null_val=df.isna().sum()

print(null_val)
```

# **Dropping the Null values**

```python
df.dropna(inplace=True)
```

# **Dropping unnecessary columns for model fitting**

```python
df.drop(['Duration in Seconds','Hashtags','Date of the Last Comment','Maximum Quality of the Video','Intern Who Collected the Data'],axis=1,inplace=True)
```

# **Changing the column datatypes for model fitting**

```python
df['Creator Gender'].replace(['Female','Male'],[0,1],inplace=True)
df['Subtitle'].replace(['No','Yes'],[0,1],inplace=True)
df['Video Description'].replace(['No','Yes'],[0,1],inplace=True)
df['Premiered or Not'].replace(['No','Yes'],[0,1],inplace=True)
```
